```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([30, 35, 40, 45, 50])
y = np.array([200, 220, 260, 280, 300])
m = len(X)
```

```python
theta0 = 0
theta1 = 0
alpha = 0.001
epochs = 10000
```

```python
def compute_cost(X, y, theta0, theta1):
    m = len(X)
    prediction = theta0 + theta1 * X
    cost = (1/(2 * m) * np.sum(prediction - y) ** 2)
    return cost
```

```python
cost_history = []

for epoch in range(epochs):
    prediction = theta0 + theta1 * X
    error = prediction - y
    
    #Calculating gradient descent 
    d_theta0 = (1/m) * np.sum(error)
    d_theta1 = (1/m) * np.sum(error * X)

    theta0 -= alpha * d_theta0
    theta1 -= alpha * d_theta1

    cost = compute_cost(X, y, theta0, theta1)
    cost_history.append(cost)

    if epoch % 100 == 0:
        print(f"Epoch {epoch}: Cost = {cost:.2f}")
```

```python
plt.plot(cost_history)
plt.xlabel("Epochs")
plt.ylabel("Cost (MSE)")
plt.title("Cost Reduction Over Time")
plt.grid(True)
plt.show()
```

```python
plt.scatter(X, y, color='blue')
predicted = theta0 + theta1 * X
plt.plot(X, predicted, color='red')
plt.xlabel("Temperature")
plt.ylabel("Revenue")
plt.title("Linear Regression with Batch Gradient Descent")
plt.grid(True)
plt.show()
```
